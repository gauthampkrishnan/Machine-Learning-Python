```
import seaborn as sns
import pandas as pd
data=pd.read_csv('train.csv')
data.head()
```

```
ax=sns.scatterplot(x="mobile_wt",y="battery_power",hue="talk_time" ,data=data)
```

```
ax=sns.distplot(data['talk_time'])
```

```
ax=sns.distplot(data['talk_time'],kde=False,rug=True,bins=20)
```

```
ax=sns.boxplot(x="four_g",y="battery_power",data=data)
```

```
ax=sns.boxplot(x="int_memory",y="price_range",data=data)
```
